```python
import numpy as np
from numpy import array
from myder import * #replacing the list of identifiers with a "*" all names defined in the module are bound, (a star imports all the functions from the file)
#from matplotlib import pyplot as plt #Matplotlib is a Python 2D plotting library 
#matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB.

import matplotlib.pyplot as plt
dx = np.linspace(0,2*np.pi,99); #np.linspace returns evenly spaced numbers over a specified interval. For ex: (start, stop, num=50) Returns num evenly spaced samples, calculated over the interval [start, stop]


#y = lambda x: np.power(np.sin(1/(x*(2-x))), 2)
#A lambda function is a small anonymous function and can take any number of arguments, but can only have one expression. Ex: lambda arguments : expression
y= lambda x: (np.sin(1/(x*(2-x))))**2
#first derivative 
dyf = forwardiff(y,2*np.pi,0,100) 
dyb = backwardiff(y,2*np.pi,0,100)
dyc = centraldiff(y,2*np.pi,0,99)
dyc3 = centraldiff(y,2*np.pi,0,99)
#print (2*np.pi)
#print (dx)
#print (dyf)
p0, = plt.plot(dx,y(dx), label= 'Origianl')
plt.legend([p0])
plt.show()
p1, = plt.plot(dx, dyf, label='Forward')
p2, = plt.plot(dx, dyb, label='Backward')
p3, = plt.plot(dx, dyc, label='Central')
p33, = plt.plot(dx,dyc3, label='Cubic')
plt.legend([p1, p2, p3, p33])
plt.show()

#Second derivative
#for Matlab file (der function) we had an option "If F is a numeric vector, it is treated as data and a simple" that took an arrays of dys and dxs values and calculated the second derivative but here we can
#The error TypeError: 'numpy.ndarray' object is not callable means that you tried to call a numpy array as a function.

#//Forward//
dyfsec = np.zeros(dyf.shape,np.float) 

dyfsec[0:-1] = np.diff(dyf)/np.diff(dx) #a[start:end] and np.diff Calculate the n-th discrete difference along the given axis.
#The first difference is given by out[n] = a[n+1] - a[n] along the given axis
dyfsec[-1] = (dyf[-1] - dyf[-2])/(dx[-1] - dx[-2]) #Negative numbers mean that you count from the right instead of the left. So, list[-1] refers to the last element, list[-2] is the second-last, and so on.
#List indexes of -x mean the xth item from the end of the list, so n[-1] means the last item in the list n

#//Backward// #//Central//
dybsec = np.zeros(dyb.shape,np.float) #.shape: The shape property is usually used to get the current shape of an array
#np.zeros: Return a new array of given shape and type, filled with zeros.
dybsec[0:-1] = np.diff(dyb)/np.diff(dx) 
dybsec[-1] = (dyb[-1]-dyb[-2])/(dx[-1]-dx[-2])
dycsec = np.zeros(dyc.shape,np.float)
dycsec[0:-1] = np.diff(dyc)/np.diff(dx)
dycsec[-1] = (dyc[-1] - dyc[-2])/(dx[-1] - dx[-2])

#//Cubic Approximation//
dycsec3 = np.zeros(dyc3.shape,np.float)
dycsec3[0:-1] = np.diff(dyc3)/np.diff(dx) 
dycsec3[-1] = (dyc3[-1]-dyc3[-2])/(dx[-1]-dx[-2])

p4, = plt.plot(dx,dyfsec, label= 'Second FD')
p5, = plt.plot(dx,dybsec, label = 'Second BD')
p6, = plt.plot(dx,dycsec, label= 'Second CD')
p66, = plt.plot(dx,dycsec3, label= 'Second Cubic')
plt.legend([p4,p5,p6,p66])
plt.show()



```
